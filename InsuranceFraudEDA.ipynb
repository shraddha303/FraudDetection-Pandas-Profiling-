## Anomaly Detection using PyCaret

```python
insurance_data["fraud_reported"].value_counts()
```

```python
from pycaret.anomaly import *
```

```python
#initialize the setup
exp_ano = setup(insurance_data)
```

```python
## creating a model
iforest=create_model('iforest')
## plotting a model
plot_model(iforest)
```

```python
## creating a model
knn=create_model('knn')
## plotting a model
plot_model(knn)
```

```python
# generate predictions using trained model
knn_predictions = predict_model(knn, data = insurance_data)
```
